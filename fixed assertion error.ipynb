```python
def to_fraction(e):
    print("Current expression:", e)
    
    if isinstance(e, tuple):
        op, left, right = e
        left_fraction = to_fraction(left)
        right_fraction = to_fraction(right)
        
        print("Operator:", op)
        print("Left fraction:", left_fraction)
        print("Right fraction:", right_fraction)
        
        if op == '/':
            if isinstance(left_fraction, tuple) and isinstance(right_fraction, tuple):
                return ('/', ('*', left_fraction, right_fraction[2]), ('*', right_fraction[1], left_fraction))
            else:
                return (op, left_fraction, right_fraction)
        elif op == '*':
            if isinstance(left_fraction, tuple) and isinstance(right_fraction, tuple):
                return ('/', ('*', left_fraction, right_fraction), ('*', 1, 1))
            else:
                return (op, left_fraction, right_fraction)
        elif op == '-':
            if isinstance(left_fraction, tuple) and isinstance(right_fraction, tuple):
                return ('/', ('-', left_fraction, right_fraction), ('*', 1, 1))
            else:
                return (op, left_fraction, right_fraction)
    else:
        return e

# Test cases
e1 = (' / ', (' / ', ' a ', 3), (' / ', 4, ' c '))
print("Expected output (Test Case 1):", ('/', ('*', 'a', 'c'), ('*', 3, 4)))
print("Output (Test Case 1):", to_fraction(e1))

e2 = (' * ', (' / ', ' a ', 3), (' / ', 4, ' c '))
print("Expected output (Test Case 2):", (' / ', (' * ', ' a ', '4'), (' * ', 3, 'c')))
print("Output (Test Case 2):", to_fraction(e2))

e3 = (' - ', (' / ', ' a ', 3), (' / ', 4, ' c '))
print("Expected output (Test Case 3):", (' / ', (' - ', (' * ', ' a ', ' c '), (' * ', ' 4 ', ' 3 ')), (' * ', 3, ' c ')))
print("Output (Test Case 3):", to_fraction(e3))
```
